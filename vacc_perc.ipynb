```python
import pandas as pd

vac = pd.read_csv('vaccinations-md.csv')
pop = pd.read_csv('populations-md.csv')
```

```python
#Merging dataframes
df = pd.merge(vac, pop, how = 'outer')

#Excerpting necessary columns, renaming the fully vaccinated column,
#and adding the PercentVaccinated column
popvac = df[['County', 'FullVaccinatedCumulative', 'Population']]
popvac = popvac.rename(columns={'FullVaccinatedCumulative': 'FullyVaccinated'})
popvac['PercentVaccinated'] = popvac['FullyVaccinated'] / popvac['Population']
popvac
```

```python
#Plotting the data
import plotly.express as px
fig = px.strip(popvac, y = 'PercentVaccinated', x = 'County',
    title = 'Vaccination Percentages By County')
fig.show()
```
